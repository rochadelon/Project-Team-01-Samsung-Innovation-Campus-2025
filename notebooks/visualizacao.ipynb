```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_ouvidoria_aneel = pd.read_csv('/workspaces/Project-Team-01-Samsung-Innovation-Campus-2025/data/df_ouvidoria_aneel_atualizado.csv')
df_ouvidoria_aneel.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_ouvidoria_aneel, x='SigUF', bins=30, kde=True)
plt.title('Distribuição de Ocorrências por Estado')
plt.xlabel('Estado (Sigla)')
plt.ylabel('Número de Ocorrências')
plt.show()
```

```python
plt.savefig('/workspaces/Project-Team-01-Samsung-Innovation-Campus-2025/images/distribuicao_ocorrencias_estado.png')
```

```python
# Anal a distribuição das categorias
categories_counts =  df_ouvidoria_aneel['NomCategoria'].value_counts()
categories_counts = pd.DataFrame(categories_counts)
print(categories_counts) 
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(data=categories_counts.reset_index(), x='NomCategoria', y='count', palette='viridis')
plt.title('Distribuição de Ocorrências por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Número de Ocorrências')
plt.xticks(rotation=45)
plt.show()
plt.savefig('/workspaces/Project-Team-01-Samsung-Innovation-Campus-2025/images/distribuicao_ocorrencias_categoria.png')
```

```python

plt.figure(figsize=(20, 10))
sns.histplot(data=categories_counts, x='NomSubCategoria', bins=100, kde=True)
plt.title('Distribuição de Ocorrências por Subcategoria')
plt.xlabel('Subcategoria')
plt.ylabel('Número de Ocorrências')
plt.show()
plt.savefig('/workspaces/Project-Team-01-Samsung-Innovation-Campus-2025/images/distribuicao_ocorrencias_subcategoria.png')
```
